```python
import matplotlib.pyplot as plt

def loss_epoch_plot(train_loss, test_loss, patient_number, lr):
    epochs = range(1, 11)  

    plt.plot(epochs, train_loss, label='Training Loss', marker='o')
    plt.plot(epochs, test_loss, label='Testing Loss', marker='o')
    plt.title(f'Patient {patient_number} - {lr}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```
